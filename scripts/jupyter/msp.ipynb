```python

# run the pipeline
def msp_relabeling(exp_name, model_file='chkpt_0000200000.pt', pref='msp0_9-200k', st=None):
    extra = ""
    if st is not None:
        extra = f"%model --switch_thresh {st}"
    %run scripts/hvs/eval_cs_offline.py --skip_video --model_file {model_file} --file data/hvsBlock3D/mode_real2_v2_fast_human_square_30k_imgs.npz --load_range 0.0 1.0 --output_file all_{pref}.npz --do_model_forward exp={exp_name} {extra}
    
    %run scripts/conversion/add_keys_to_npz.py --y --file data/hvsBlock3D/human_square_30k.npz --keys_file {exp_name}/all_{pref}.npz --keys evaluations/click_state --keys_out click_state --no_overwrite --output_file data/hvsBlock3D/raw_mode_{pref}_human_square_30k.npz
    
    %run scripts/hvs/relabel_by_mode.py --y --file data/hvsBlock3D/raw_mode_{pref}_human_square_30k.npz --output_file data/hvsBlock3D/mode_{pref}_fast_human_square_30k.npz --populate_mode0_actions --fill_n_mode0_action_dims 6 cfgs/exp_hvs/square/hydra.py %dataset_in cfgs/dataset/np_seq.py %dataset_out cfgs/dataset/np_seq.py

# run the pipeline
def msp_video(exp_name, model_file='chkpt_0000200000.pt', pref='msp0_9-200k', st=None):
    extra = ""
    if st is not None:
        extra = f"%model --switch_thresh {st}"
    %run scripts/hvs/eval_cs_offline.py --model_file {model_file} --file data/hvsBlock3D/mode_real2_v2_fast_human_square_30k_imgs.npz --load_range 0.9 1.0 --output_file last_{pref}.npz --do_model_forward exp={exp_name} %env_train --imgs {extra}

def play_video(file):
    from IPython.display import Video
    return Video(file, embed=True)
```
